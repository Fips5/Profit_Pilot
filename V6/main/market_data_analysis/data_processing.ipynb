```python
# Load the LSTM model
news_model_path = r'C:\Users\David\Documents\ProfitPilot\V6\tests\TEST_1\models\model-1\model-1.h5'
news_lstm_model = load_model(news_model_path)
```

```python
nvda_news_df = get_latest_news_df('NVDA')
```

```python
tokenizer =  tokenizer_return()
```

```python
nvda_score_col = []
nvda_text_col = []
for link in nvda_news_df['URL']:
    article_text = extract_article_text(link)
    processed_input = input_processing(article_text, tokenizer, maxlen = 100)
    prediction = news_lstm_model.predict(processed_input)
    nvda_score_col.append(prediction)
    nvda_text_col.append(article_text)
nvda_score_col = [float(arr[0][0]) for arr in nvda_score_col]
```

```python
results_df = pd.DataFrame({
    'Text': nvda_text_col,
    'Score': nvda_score_col
})
results_df
```
