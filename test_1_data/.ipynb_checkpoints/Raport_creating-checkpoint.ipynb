```python
from ib_insync import IB, Fill
import pandas as pd
from datetime import datetime, timedelta
import asyncio
import time
class IBKRClient:
    def __init__(self, host='127.0.0.1', port=7496, client_id=1):
        self.ib = IB()
        self.host = host
        self.port = port
        self.client_id = client_id

    async def connect(self):
        if not self.ib.isConnected():
            await self.ib.connectAsync(self.host, self.port, clientId=self.client_id)

    async def get_fills(self, days_back=1):
            async def get_fills(self, days_back=1):
        await self.connect()
        fills: list[Fill] = self.ib.fills()
        print("🔍 Raw fills returned by IBKR:")
        for f in fills:
            print(f)

        cutoff = datetime.now() - timedelta(days=days_back)
        filtered = [
            {
                'Time': f.execution.time.strftime('%Y-%m-%d %H:%M:%S'),
                'Symbol': f.contract.symbol,
                'Action': f.execution.side,
                'Quantity': f.execution.shares,
                'Price': f.execution.price,
                'Commission': f.execution.commission,
                'Order ID': f.execution.orderId,
                'Exchange': f.execution.exchange
            }
            for f in fills if f.execution.time.replace(tzinfo=None) >= cutoff
        ]

        print(f"✅ Filtered {len(filtered)} fills after date cutoff")
        return pd.DataFrame(filtered)


# Notebook-compatible async run
async def run_ibkr_export():
    client = IBKRClient()
    df = await client.get_fills(days_back=30)
    csv_path = 'ibkr_daily_trades.csv'
    df.to_csv(csv_path, index=False)
    print(df)
    print(f"✅ IBKR trades saved to {csv_path}")
    client.disconnect()

# Run in Jupyter or other async environment
await run_ibkr_export()
```
